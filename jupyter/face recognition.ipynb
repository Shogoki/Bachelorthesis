## running haarcascade face detection (pretrained)
### loading first frame 

```python

image = cv2.imread("../data/extracted/frame1.jpg")
image
```

```python

detection_model_path = '../haarcascade_files/haarcascade_frontalface_default.xml'

face_detection = cv2.CascadeClassifier(detection_model_path)

faces = face_detection.detectMultiScale(image,scaleFactor=1.1,minNeighbors=5,minSize=(30,30),flags=cv2.CASCADE_SCALE_IMAGE)

faces
```

```python
for face in faces:
    (fX, fY, fW, fH) = face
    cut = image[ fY: fY + fH, fX: fX +fW].copy()
    cv2.rectangle(image, (fX, fY), (fX+ fW, fY + fH), 20)

plt.imshow(image)
plt.show()
```

```python
plt.imshow(cut)
plt.show()
```

now making it grayscale:


```python
grayface =cv2.cvtColor(cut, cv2.COLOR_BGR2GRAY)

plt.imshow(grayface)
plt.show()
grayface.shape
```

```python
grayface = cv2.resize(grayface, (48, 48))

plt.imshow(grayface)
plt.show()
grayface.shape
```
